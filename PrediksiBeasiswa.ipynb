## Load Data

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
from matplotlib import rcParams
import warnings
```

```python
rcParams['figure.figsize'] = 15, 8
mpl.style.use(['ggplot'])
warnings.filterwarnings("ignore")
```

Dataset Federal University of Rio Grande do Sul, Brazil Entrance Test Score and GPA 

Sumber : https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/O35FW8

```python
df_brazil = pd.read_csv('https://drive.google.com/uc?id=1Co_kFe4tOpGEXPEe1pGsa6adi0A3zRtq')
df_brazil.head()
```

Dataset Educational Testing Service berisi SAT (USA Entrance Exam), GPA High School, GPA First Year College

Sumber : https://www.openintro.org/data/index.php?data=satgpa

```python
df_usa = pd.read_csv('https://drive.google.com/uc?id=1KUPYuoN8uGDG5ZI2WV0dNVXGbRP7h6B0')
df_usa.head()
```

## Data Exploration Brazil

```python
df_brazil.shape
```

```python
df_brazil.info()
```

```python
df_brazil.describe()
```

```python
kolom_brazil = list(df_brazil.columns)
kolom_brazil
```

```python
for i in range(0, len(kolom_brazil)):
    plt.subplot(1, len(kolom_brazil), i+1)
    sns.boxplot(y=df_brazil[kolom_brazil[i]], color='gray', orient='v')
    plt.tight_layout()
```

```python
for i in range(0, len(kolom_brazil)):
    plt.subplot(3, 4, i+1)
    sns.distplot(df_brazil[kolom_brazil[i]], color='gray')
    plt.tight_layout()
```

```python
plt.figure(figsize=(12, 12))
sns.heatmap(df_brazil.corr(), cmap='Blues', annot=True, fmt='.2f')
```
